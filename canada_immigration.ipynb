6) Using a pie chart, explore the proportion (percentage) of new immigrants grouped by continents in the year 2013

```python
# creating new data set df_continents
df_continents = df.drop(columns=['Type','Coverage','AREA', 'OdName','REG','RegName','DEV','DevName'])
df_continents
```

```python
# creating new column which contain sum of all data from 1980-2013 of each row
df_continents['total']=df_continents.iloc[: , 1:].sum(axis=1)
```

```python
df_continents
```

```python
# using grouby method 1st group the continents and then appling sum to total column to get total immigrant of each continent
df_continents =df_continents.groupby(['AreaName'])['total'].sum()
```

```python
df_continents
```

```python
# using pie chart we can vizualize the % wise contribution of immigrant from each continent
plt.figure(figsize=(14,8))
X = df.columns[9:42].tolist()
labels = ['Africa','Asia', 'Europe','Latin America and the Caribbean', 'Northern America','Oceania']
immigrant_2013 = df_continents.tolist()
plt.axis("equal")
plt.pie(immigrant_2013, labels=labels, autopct='%1.1f%%')
plt.title('2013 Immigration Data')
```

10) Let's start by analyzing the effect of Argentina's great depression.

Argentina suffered a great depression from 1998 - 2002, which caused widespread unemployment, riots, the fall of the government, and a default on the country's foreign debt. In terms of income, over 50% of Argentines were poor, and seven out of ten Argentine children were poor at the depth of the crisis in 2002.

Let's analyze the effect of this crisis, and compare Argentina's immigration to that of it's neighbour Brazil. Let's do that using a bubble plot of immigration from Brazil and Argentina for the years 1980 - 2013. We will set the weights for the bubble as the normalized value of the population for each year

```python
canada80_2013
```

```python
Argentina = canada80_2013['Argentina']
Brazil = canada80_2013['Brazil']
```

```python
# normalizing the values 
a_norm =Argentina / Argentina.max()
b_norm =Brazil/ Brazil.max()
```

```python
# bubble plot
plt.figure(figsize=(14,8))
plt.scatter(year, Argentina, color='pink', alpha= 0.5, s= a_norm * 2000)
plt.scatter(year, Brazil, color='green', alpha= 0.5, s= b_norm * 2000)
plt.xlabel('Years', size=14)
plt.ylabel('Number of immigrants',size=14)
```

Brazil suffered the Samba Effect where the Brazilian real (currency) dropped nearly 35% in 1999. There was a fear of a South American financial crisis as many South American countries were heavily dependent on industrial exports from Brazil. The Brazilian government subsequently adopted an austerity program, and the economy slowly recovered over the years, culminating in a surge in 2010.
